### Snake: Simulate a snake game and print the game results.

You are given a map that ‘x’ represents a rock, ‘-’represents a space, ‘#’represents the body of snake. ‘@’represent the head of the snake and a sequence of actions that ‘0,1,2,3’represent to move to up/down/left/right correspondingly for one step.
A greedy snake starts in the map state and moves one step per unit of time according to the sequence of actions until all actions complete or fail. It will fail when the head and the stone overlap, the head goes beyond the boundary, or the head overlaps the body. 

#### Input
A matrix with type char (the map). 
A sequence with type int (the motions). 

#### Output
the the result of the game:
If it failed, output the running time of the game.
It it didn’t fail, output the final position of the head (in the form “%d, %d”).

```python
"""
Example:
input:
map:
---------
------x--
-x-------
---@-----
---##----
------x--
--x----x-
-x-------
---------
action:
0 0 3 3 0 3 3 1 1 1 1 1 3 1 1 2 2 2 2 2

output:
7 3
"""
```

```python
# add your code here
def show_map(map):
    for row in map:
        print(''.join(row))

def move(map, action):
    headx, heady = 0, 0
    for i in range(len(map)):
        for j in range(len(map[i])):
            if map[i][j] == '@':
                headx, heady = i, j

    snake = [(headx, heady)]
    last = (headx, heady)
    current = (headx, heady)
    while True:
        if map[current[0] + 1][current[1]] == '#' and current[0] + 1 != last[0]:
            snake.append((current[0] + 1, current[1]))
            last = current
            current = (current[0] + 1, current[1])
        elif map[current[0] - 1][current[1]] == '#' and current[0] - 1 != last[0]:
            snake.append((current[0] - 1, current[1]))
            last = current
            current = (current[0] - 1, current[1])
        elif map[current[0]][current[1] + 1] == '#' and current[1] + 1 != last[1]:
            snake.append((current[0], current[1] + 1))
            last = current
            current = (current[0], current[1] + 1)
        elif map[current[0]][current[1] - 1] == '#' and current[1] - 1 != last[1]:
            snake.append((current[0], current[1] - 1))
            last = current
            current = (current[0], current[1] - 1)
        else:
            break

    flag = True
    for i in range(len(action)):
        if action[i] == 0:
            if headx != 0 and (map[headx - 1][heady] == '-' or map[headx - 1][heady] == '#' and (headx - 1, heady) == snake[-1]):
                map[headx][heady] = '#'
                map[snake[-1][0]][snake[-1][1]] = '-'
                headx -= 1
                map[headx][heady] = '@'
                snake.insert(0, (headx, heady))
                snake.pop()
            else:
                flag = False
        elif action[i] == 1:
            if headx != len(map) - 1 and (map[headx + 1][heady] == '-' or map[headx + 1][heady] == '#' and (headx + 1, heady) == snake[-1]):
                map[headx][heady] = '#'
                map[snake[-1][0]][snake[-1][1]] = '-'
                headx += 1
                map[headx][heady] = '@'
                snake.insert(0, (headx, heady))
                snake.pop()
            else:
                flag = False
        elif action[i] == 2:
            if heady != 0 and (map[headx][heady - 1] == '-' or map[headx][heady - 1] == '#' and (headx, heady - 1) == snake[-1]):
                map[headx][heady] = '#'
                map[snake[-1][0]][snake[-1][1]] = '-'
                heady -= 1
                map[headx][heady] = '@'
                snake.insert(0, (headx, heady))
                snake.pop()
            else:
                flag = False
        elif action[i] == 3:
            if heady != len(map[0]) - 1 and (map[headx][heady + 1] == '-' or map[headx][heady + 1] == '#' and (headx, heady + 1) == snake[-1]):
                map[headx][heady] = '#'
                map[snake[-1][0]][snake[-1][1]] = '-'
                heady += 1
                map[headx][heady] = '@'
                snake.insert(0, (headx, heady))
                snake.pop()
            else:
                flag = False
        if flag:
            print('Move {}: {}'.format(i+1, action[i]))
            print('snake: {}'.format(snake))
            show_map(map)
        else:
            print('Game over, {} actions in total'.format(i+1))
            return
        
    print('Final position of head: {} {}'.format(headx, heady))

                
            
```
